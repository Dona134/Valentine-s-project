## Proportion of Messages in English and Russian

```python
from langdetect import detect, LangDetectException
```

```python
def detect_language(text):
    try:
        # Check if the text is a string and not empty
        if isinstance(text, str) and len(text) >= 3:
            return detect(text)
        else:
            return 'ru'  # Default to Russian for non-string or short texts
    except LangDetectException:
        # If detection fails, label as Russian
        return 'ru'
```

```python
# Sample a fraction of your data, e.g., 10%
sampled_messages = messages.sample(frac=0.1)

# Apply the function to the sample
sampled_messages['lang'] = sampled_messages['text'].apply(detect_language)
```

```python
english_count = len(sampled_messages[sampled_messages['lang'] == 'en'])
russian_count = len(sampled_messages[sampled_messages['lang'] == 'ru'])

# Total number of messages
total_count = len(sampled_messages)

# Calculate proportions
english_prop = english_count / total_count
russian_prop = russian_count / total_count

# Data for pie chart
labels = 'English', 'Russian'
sizes = [english_prop, russian_prop]
```

```python
# Plotting pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Proportion of Messages in English and Russian')
plt.show()
```

## Calculating average response time

```python

# Calculate the time difference between each message and the previous one
messages['response_time'] = messages['date'].diff()

# Convert the time difference to a suitable format, e.g., seconds
messages['response_time_seconds'] = messages['response_time'].dt.total_seconds()

# Determine the sender of the previous message
messages['previous_from'] = messages['from'].shift()

# Filter out rows where the sender is the same as the previous message
df_filtered = messages[messages['from'] != messages['previous_from']]

# Group by sender and calculate the average response time
average_response_time = df_filtered.groupby('from')['response_time_seconds'].mean()

print(average_response_time/60)
messages.head()
```

## top 5 most used emojis for each sender

```python
from collections import defaultdict, Counter
```

```python
#  This is a special dictionary where each key is a sender's name, and the value is another dictionary (a Counter) that keeps track of how many times each emoji is used by that sender.
emoji_counts_per_sender = defaultdict(Counter)

# Function to extract emojis from a text
def extract_emojis(text):
    text_as_str = str(text)
    return emoji.distinct_emoji_list(text_as_str)
```

```python
# Iterate over each row in the DataFrame
for index, row in df_filtered_emojis.iterrows():
    sender = row['from']
    message = row['text']
    emojis = extract_emojis(message)
    for em in emojis:
        emoji_counts_per_sender[sender][em] += 1
```

```python
top_5_emojis_per_sender = {sender: counts.most_common(5) for sender, counts in emoji_counts_per_sender.items()}
```

```python
# Display the results
for sender, emojis in top_5_emojis_per_sender.items():
    print(f"Sender: {sender}")
    for em, count in emojis:
        print(f"Emoji: {em}, Count: {count}")
    print()
```

## Word Cloud Generation

```python
from wordcloud import WordCloud
import nltk
```

```python
# Download Russian stopwords
nltk.download('stopwords')

# Import Russian stopwords from nltk
from nltk.corpus import stopwords
russian_stopwords = set(stopwords.words('russian'))
```

```python
# Concatenate all messages into one text
all_text = ' '.join(messages['text'].astype(str))

# Generate Word Cloud
wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white', 
                      min_font_size = 10).generate(all_text)

# Plotting
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

### Let's exlude some basic russian words (stopwords) and 10 most common words 

```python
from wordcloud import WordCloud, STOPWORDS




# Create a set for stopwords
stopwords = set(STOPWORDS)
stopwords.update(russian_stopwords)


# Generate Word Cloud excluding the top 30 most common words
wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white', 
                      stopwords = stopwords,
                      min_font_size = 10).generate(all_text)

# Plotting
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
